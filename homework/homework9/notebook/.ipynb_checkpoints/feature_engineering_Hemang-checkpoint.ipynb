## TODO: Implement at least 2 engineered features here

```python
# One-hot encode the marital status column
df = pd.get_dummies(df, columns=['Marital status'], prefix='Marital')
```

### Rationale for Feature 1

Marital status often reflects important social and economic factors that can indirectly influence a student’s academic performance. For example, married students might have different time commitments or family support compared to single students, affecting study time or stress levels. One-hot encoding this categorical feature converts it into a numerical format suitable for machine learning algorithms without implying any rank or order. This preserves all category distinctions, allowing the model to learn distinct patterns related to each marital status group.

```python
df['pass_rate_1st_sem'] = df['Curricular units 1st sem (approved)'] / df['Curricular units 1st sem (enrolled)']
```

### Rationale for Feature 2

The pass rate, calculated as the ratio of approved curricular units to those enrolled in the first semester, is a strong direct indicator of academic success and persistence. Students with higher pass rates demonstrate better course completion which often correlates with overall academic achievement (the target). This feature synthesizes detailed semester data into a simple, interpretable metric that can highlight students at risk or likely to perform well. Such composite features typically improve prediction accuracy by capturing holistic academic behavior trends compared to raw counts alone.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot of pass_rate_1st_sem versus Target
plt.figure(figsize=(8, 6))
sns.scatterplot(x='pass_rate_1st_sem', y='Target', data=df)
plt.title('Pass Rate 1st Semester vs Target')
plt.xlabel('Pass Rate 1st Semester')
plt.ylabel('Target')
plt.show()
```
